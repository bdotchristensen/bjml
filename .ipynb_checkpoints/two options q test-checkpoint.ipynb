```python
import torch
import random
import torch.nn as nn
import matplotlib.pyplot as plt

from torch.optim import RMSprop
from torch.utils.data import DataLoader, TensorDataset
```

```python
class QFunction(nn.Module):
    def __init__(self):
        super().__init__()
        self.lin1 = nn.Linear(2, 12)
        self.lin2 = nn.Linear(12, 12)
        self.lin3 = nn.Linear(12, 12)
        self.lin4 = nn.Linear(12, 1)
        
    def forward(self, x):
        x = self.lin1(x)
        x = torch.relu(x)
        x = self.lin2(x)
        x = torch.relu(x)
        x = self.lin3(x)
        x = torch.relu(x)
        return self.lin4(x)

class ReplayMemory(nn.Module):
    def __init__(self, N = 200, mini_batch = 100):
        self.memory = []
        self.N = N
        self.mini_batch = mini_batch

    def push(self, tup):
        self.memory.append(tup)
        
        if len(self.memory) > self.N:
            self.memory = self.memory[1:]

    def sample(self):
        return random.sample(self.memory, min(self.memory.__len__(), self.mini_batch))
```

```python
EPOCHS = 10000
Q = QFunction()
D = ReplayMemory()

optimizer = RMSprop(Q.parameters(), lr = 0.001)
epsilon = 0.05

for epoch in range(EPOCHS):
    # state -- between 0 and 1. denotes probability of winning
    state = torch.rand(1, 1)

    action0 = Q(torch.cat((state, torch.zeros(1, 1)), axis = -1)).detach().item()
    action1 = Q(torch.cat((state, torch.ones(1, 1)), axis = -1)).detach().item()

    if torch.rand(1) < epsilon:
        action = [action0, action1].index(min(action0, action1)) # not greedy

    else:
        action = [action0, action1].index(max(action0, action1)) # greedy
    
    
    # get reward
    x = torch.bernoulli(state).item()
    reward = (2 * x - 1) * (2 * action - 1)

    # push to memory
    SARS = (state, action, reward, torch.zeros(1, 1))
    D.push(SARS)

    # step in the q function 
    sample = D.sample()
        
    phi_j = torch.vstack([x[0] for x in sample])
    a_j = torch.tensor([x[1] for x in sample]).reshape(-1, 1)
    r_j = torch.tensor([x[2] for x in sample]).reshape(-1, 1)

    input = torch.hstack((phi_j, a_j))

    optimizer.zero_grad()

    loss = torch.sum((r_j - Q(input))**2)
    loss.backward()

    optimizer.step()

    # show learned Q function outputs
    if epoch % 700 == 0:
        p = torch.linspace(0, 1, 200).reshape(200, 1)
        
        zeros = torch.zeros(200, 1)
        input = torch.hstack((p, zeros))
        
        with torch.no_grad():
            print('=== Epoch', epoch, '===')
            plt.plot(p, Q(input))

        ones = torch.ones(200, 1)
        input = torch.hstack((p, ones))
        with torch.no_grad():
            plt.plot(p, Q(input))
            plt.show()
        
    
```
