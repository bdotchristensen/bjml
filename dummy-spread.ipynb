```python
import torch
import torch.nn as nn
from torch.optim import Adam
from torch.utils.data import DataLoader, TensorDataset

import random
import matplotlib.pyplot as plt
from torch.optim import Adam
```

```python
class MonotonicBetSpread(nn.Module):
    def __init__(self, spread = 6):
        super().__init__()
        self.lin1 = nn.Linear(1, 12) 
        self.lin2 = nn.Linear(12, 12)
        self.lin3 = nn.Linear(12, 1)

    def forward(self, x):

        self.lin1.weight = nn.Parameter(torch.relu(self.lin1.weight))
        self.lin2.weight = nn.Parameter(torch.relu(self.lin2.weight))
        self.lin3.weight = nn.Parameter(torch.relu(self.lin3.weight))
        
        x = self.lin1(x)
        x = torch.relu(x)
        x = self.lin2(x)
        x = torch.relu(x)
        x = self.lin3(x)
        
        return x


better = MonotonicBetSpread()
optimizer = Adam(better.parameters(), lr = 0.01)

x = torch.arange(-15, 16).reshape(-1, 1) * 1.
y = (torch.arange(-15, 16).reshape(-1, 1) / 4)**2
```

```python
losses = []

for i in range(10000):
    optimizer.zero_grad()

    loss = torch.mean((better(x) - y)**2) # 10 * better.monotonic_loss()**2 + 
    loss.backward()

    losses.append(loss.item())

    optimizer.step()

    if i % 500 == 0:
        plt.plot(x, y)
        plt.plot(x, better(x).detach())
        plt.show()
        
        print(loss.item()) #, better.lin1.bias)
```

```python
plt.plot(losses)
```
